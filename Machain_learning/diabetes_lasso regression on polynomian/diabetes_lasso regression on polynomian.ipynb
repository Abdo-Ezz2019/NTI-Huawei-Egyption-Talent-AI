```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.preprocessing import PolynomialFeatures
```

```python
df=pd.read_csv('diabetes.csv')
df.info()
```

```python
x=df.drop('Outcome',axis=1)
y=df.Outcome
print(x.shape,y.shape)
```

```python

x_df = pd.DataFrame(x)
y_df = pd.DataFrame(y)

print("DataFrame for X:")
print(x_df.head())

print("\nDataFrame for Y:")
print(y_df.head())
```

```python
y=df['Outcome']
x=df.drop('Outcome',axis=1) # to remove the coloum if u want to remove raw = 0
x
y
```

```python
# train test
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
x_train
```

```python
model_1 = linear_model.LinearRegression()
model_1.fit(x_train, y_train)
```

```python

y_pred_train = model_1.predict(x_train)
print("MSN for Train Data:", mean_squared_error(y_train, y_pred_train))
```

```python
print("r2 for train data",r2_score(y_train,y_pred_train) )
```

```python
y_pred = model_1.predict(x_test)
print("MSN for Test Data:", mean_squared_error(y_test, y_pred))
```

```python
print("r2 for test data",r2_score(y_test,y_pred) )
```
