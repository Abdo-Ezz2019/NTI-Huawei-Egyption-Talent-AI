```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
```

```python
housing = pd.read_csv('/content/housing.csv', delim_whitespace=True)
housing
```

```python
y = housing.iloc[:, -1]
X = housing.iloc[:, :-1]
X
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
X = X_scaled_df
X
```

```python
X.isnull().sum()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train
```

```python
model_1 = linear_model.LinearRegression()
model_1.fit(X_train, y_train)
```

```python
y_pred_train = model_1.predict(X_train)
mean_squared_error(y_train, y_pred_train)
r2_score(y_train, y_pred_train)
```

```python


poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
X_train_poly, X_test_poly, y_train, y_test = train_test_split(X_poly, y, test_size=0.3, random_state=42)
model_poly = LinearRegression()
model_poly.fit(X_train_poly, y_train)
y_pred_train_poly = model_poly.predict(X_train_poly)
print("Polynomial Regression (degree 2) - Training MSE:", mean_squared_error(y_train, y_pred_train_poly))
print("Polynomial Regression (degree 2) - Training R2:", r2_score(y_train, y_pred_train_poly))
y_pred_test_poly = model_poly.predict(X_test_poly)
print("Polynomial Regression (degree 2) - Test MSE:", mean_squared_error(y_test, y_pred_test_poly))
print("Polynomial Regression (degree 2) - Test R2:", r2_score(y_test, y_pred_test_poly))
```

```python
lasso_model = linear_model.Lasso(alpha=0.03)
lasso_model.fit(X_train_poly, y_train)
y_pred_train_lasso = lasso_model.predict(X_train_poly)
print("\nLasso Regression on Polynomial Features - Training MSE:", mean_squared_error(y_train, y_pred_train_lasso))
print("Lasso Regression on Polynomial Features - Training R2:", r2_score(y_train, y_pred_train_lasso))
y_pred_test_lasso = lasso_model.predict(X_test_poly)
print("Lasso Regression on Polynomial Features - Test MSE:", mean_squared_error(y_test, y_pred_test_lasso))
print("Lasso Regression on Polynomial Features - Test R2:", r2_score(y_test, y_pred_test_lasso))
```
