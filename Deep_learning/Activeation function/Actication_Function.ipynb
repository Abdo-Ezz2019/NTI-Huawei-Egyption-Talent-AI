```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math

```

```python
# my sigmode function
def my_sigmode(x):
  return 1/(1+np.exp(-x))
# function tanh
def my_tanh(x):
  return np.tanh(x)
# softmax function
def my_softmax(x):
  return np.exp(x)/np.sum(np.exp(x),axis=0)
# Relu
def my_relu(x):
  return np.maximum(0,x)
# Leaky Relu
def my_leaky_relu(x):
  return np.maximum(0.01*x,x)
def my_swish(x):
  return x/(1+np.exp(-x))

```

```python
x=np.arange(-10,10,0.1)
y=my_sigmode(x)
y_bar=y*(1-y)  # to make Pi
plt.plot(x,y)
plt.plot(x,y_bar)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Sigmode")
plt.grid()
plt.show()
```

```python
x=np.arange(-10,10,0.1)
y=my_tanh(x)
y_bar=(1-y**2)
plt.plot(x,y)
plt.plot(x,y_bar)
plt.xlabel("x")
plt.ylabel("y")
plt.title("tanh")
plt.grid()
plt.show()
```

```python
x=np.arange(-5,5,0.1)
y=my_softmax(x)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("softmax")
plt.grid()
plt.show()
```
